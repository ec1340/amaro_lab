### Title of my project in the Amaro Lab:

# Emperically-Derived Heuristics for Simulation-Enabled Estimation of Kinetic Rates


### or

# Finding out how to get good kinetic values from simulations

# Motivation

## Docking is fast at the expense of accuracy

<img src="docking.png",width=700,height=700>

## It also only gives a thermodynamic (delta G of binding) estimate of how well the ligand will bind

## Research has shown that kinetic values (k-on and k-off) are better indicators of in-vivo efficacy

<img src="kinetics.png",width=500,height=500>

## To determine these kinetic properties from simulations, a lot of simulation data must be accumulated


## Binding and unbinding events can occur on timescales infeasible to simulate

## Goal:

# To derive k-on and k-off values using milestoning theory

# Milestoning Theory

## We can determine k-on and k-off by simulating individual regions of space one at a time 

<img src="milestoning.png",width=500,height=500>

# How can we do this?



## We simulate each region and record how often the drug moves from one region to the next


## We collect these transition statisitics as well as how long it takes for each transition to occur

## We can then use this data to calculate k-on and k-off rates

# But this has been done before,
<img src="elber.png",width=700,height=700>

# Our method: Integrate Molecular Dynamics and Brownian Dynamic Simulations Using Milestoning theory
